```python
x_1 = tf.constant([5,3,2,4,5,8,9] ,dtype=tf.float32)
x_2 = tf.constant([5,3,2,4,5,0,9] ,dtype=tf.float32)

print(tf.math.divide_no_nan(x_1,x_2))

```

```python
x_1 = tf.constant([5,3,2,4,5,8,9] ,dtype=tf.float32)
x_2 = tf.constant([[4],[5],[5]] ,dtype=tf.float32)

print(tf.math.maximum(x_1,x_2))


```

## **Ragged** **Tensors**

```python
tensor_two_d = [[1,2,0],
                [3,],
                [1,5,6,5,6],
                [2,3]]

tensor_ragged = tf.ragged.constant(tensor_two_d)
print(tensor_ragged)
```

## Sparse Tensors


```python
tensor_sparse = tf.sparse.SparseTensor(
    indices=[[1,1], [3,4]],
    values=[11, 23],
    dense_shape=[5, 6]
)
print(tensor_sparse)
```

```python
tf.sparse.to_dense(tensor_sparse)
```
